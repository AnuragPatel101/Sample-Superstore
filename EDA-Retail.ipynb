### Ship Mode

```python
data['Ship Mode'].value_counts().plot(kind = 'bar')
plt.title('Total transactions for Ship Mode')
plt.show()
```

```python
data.groupby('Ship Mode')['Profit'].sum().plot(kind = 'bar')
plt.title('Total Profit for Ship Mode')
plt.show()
```

```python
data.groupby('Ship Mode')['Profit'].mean().plot(kind = 'bar')
plt.title('Average Profit for Ship Mode')
plt.show()
```

```python
data.groupby('Ship Mode')['Quantity'].sum().plot(kind = 'bar')
plt.title('Total Quantity for Ship Mode')
plt.show()
```

```python
data.groupby('Ship Mode')['Sales'].sum().plot(kind = 'bar')
plt.title('Total Sales for Ship Mode')
plt.show()
```

```python
data.groupby('Ship Mode')['Sales'].mean().plot(kind = 'bar')
plt.title('Average Sales for Ship Mode')
plt.show()
```
